## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from datetime import datetime

# Config
DATASET_FOLDER = 'dataset'
RFM_WEIGHTS = {'Monetary': 5, 'Frequency': 5, 'Recency': 2}
SULTAN_QUANTILE = 0.90  # top 10% by weighted RFM score
```

## Data Wrangling

### Gathering Data

```python
# Load all datasets
orders_df = pd.read_csv(f'{DATASET_FOLDER}/orders_dataset.csv')
order_items_df = pd.read_csv(f'{DATASET_FOLDER}/order_items_dataset.csv')
order_payments_df = pd.read_csv(f'{DATASET_FOLDER}/order_payments_dataset.csv')
customers_df = pd.read_csv(f'{DATASET_FOLDER}/customers_dataset.csv')
geo_df = pd.read_csv(f'{DATASET_FOLDER}/geolocation_dataset.csv')
products_df = pd.read_csv(f'{DATASET_FOLDER}/products_dataset.csv')
product_category_name_translation_df = pd.read_csv(f'{DATASET_FOLDER}/product_category_name_translation.csv')
sellers_df = pd.read_csv(f'{DATASET_FOLDER}/sellers_dataset.csv')

# Quick dataset sanity check
print('orders_df:', orders_df.shape)
print('order_items_df:', order_items_df.shape)
print('customers_df:', customers_df.shape)
print('geo_df:', geo_df.shape)
print('products_df:', products_df.shape)
```

### Cleaning Data

```python
# Cleaning steps
order_items_df = order_items_df.copy()
orders_df = orders_df.copy()

# 1) Outlier handling for item price (keep consistent with existing notebook logic)
q1 = order_items_df['price'].quantile(0.25)
q3 = order_items_df['price'].quantile(0.75)
iqr = q3 - q1
price_upper = 1.5 * iqr
order_items_df = order_items_df[order_items_df['price'] <= price_upper].copy()
print('order_items_df after price filter:', order_items_df.shape)

# 2) Parse datetime fields
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'], errors='coerce')
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'], errors='coerce')
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'], errors='coerce')
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'], errors='coerce')
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'], errors='coerce')

# 3) Keep orders with a valid approval timestamp (needed for time series + RFM)
orders_df = orders_df.dropna(subset=['order_approved_at']).copy()
print('orders_df after dropna(order_approved_at):', orders_df.shape)

# 4) Build merged fact table
order_merge = order_items_df.merge(orders_df, on='order_id', how='inner')
order_merge = order_merge.dropna(subset=['order_approved_at', 'price', 'customer_id']).copy()

# Convenience time fields
order_merge['year'] = order_merge['order_approved_at'].dt.year
order_merge['month'] = order_merge['order_approved_at'].dt.month
order_merge['year_month'] = order_merge['order_approved_at'].dt.to_period('M').astype(str)

print('order_merge shape:', order_merge.shape)
print('order_merge columns (subset):', [c for c in ['order_id','customer_id','product_id','price','order_approved_at','year','month','year_month'] if c in order_merge.columns])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Q1) Kenaikan Revenue dari tahun ke tahun

yearly_revenue = (
    order_merge.groupby('year', as_index=False)
    .agg(revenue=('price', 'sum'), orders=('order_id', 'nunique'))
    .sort_values('year')
)

yearly_revenue['yoy_growth_%'] = yearly_revenue['revenue'].pct_change() * 100

a_best = yearly_revenue.loc[yearly_revenue['yoy_growth_%'].idxmax()]
print('Yearly revenue & YoY growth (%)')
print(yearly_revenue.to_string(index=False))
print('\nTahun dengan kenaikan YoY terbesar:')
print(a_best[['year','revenue','yoy_growth_%']].to_string())

plt.figure(figsize=(9, 4))
sns.lineplot(data=yearly_revenue, x='year', y='revenue', marker='o')
plt.title('Revenue per Tahun (berdasarkan sum(price))')
plt.xticks(yearly_revenue['year'].unique())
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
# Q2) Dibulan apa penjualan cenderung meningkat?
# - Revenue per bulan kalender = sum(price)
# - “Cenderung meningkat” diproksikan dengan pola MoM (month-over-month) % growth

monthly_revenue = (
    order_merge.groupby(['year_month', 'year', 'month'], as_index=False)
    .agg(revenue=('price', 'sum'))
    .sort_values('year_month')
)

monthly_revenue['mom_growth'] = monthly_revenue['revenue'].diff()
monthly_revenue['mom_growth_%'] = monthly_revenue['revenue'].pct_change() * 100

# Aggregate MoM trend by calendar month number
month_trend = (
    monthly_revenue.dropna(subset=['mom_growth_%'])
    .groupby('month', as_index=False)
    .agg(
        avg_mom_growth_%=('mom_growth_%', 'mean'),
        positive_mom_rate=('mom_growth_%', lambda s: (s > 0).mean()),
        n_months=('mom_growth_%', 'size'),
    )
)

month_name_id = {
    1:'Januari',2:'Februari',3:'Maret',4:'April',5:'Mei',6:'Juni',
    7:'Juli',8:'Agustus',9:'September',10:'Oktober',11:'November',12:'Desember'
}

# Pick the “most increasing” month number by: highest positive rate, then highest avg growth
month_trend = month_trend.sort_values(['positive_mom_rate','avg_mom_growth_%'], ascending=False)
best_month_row = month_trend.iloc[0]
best_month_num = int(best_month_row['month'])
print('Ringkasan tren MoM per bulan kalender:')
print(month_trend.to_string(index=False))

print('\nBulan yang cenderung meningkat (berdasarkan positive MoM rate tertinggi, lalu avg growth):')
print(f"- {best_month_num} ({month_name_id[best_month_num]}): positive_mom_rate={best_month_row['positive_mom_rate']:.3f}, avg_mom_growth_%={best_month_row['avg_mom_growth_%']:.2f}")

# Plot monthly revenue over time
plt.figure(figsize=(10, 4))
sns.lineplot(data=monthly_revenue, x='year_month', y='revenue', marker='o')
plt.xticks(rotation=45, ha='right')
plt.title('Revenue per Bulan (sum(price))')
plt.tight_layout()
plt.show()

# Plot month-of-year trend
plt.figure(figsize=(10, 4))
sns.barplot(data=month_trend.sort_values('month'), x='month', y='avg_mom_growth_%')
plt.xticks(range(0, 12), [month_name_id[i] for i in range(1, 13)], rotation=45)
plt.title('Rata-rata MoM Growth (%) per Bulan Kalender')
plt.tight_layout()
plt.show()
```

## Metodologi Lengkap (Q1–Q5)

Catatan definisi umum:
- **Revenue** dihitung sebagai **jumlah `price`** pada dataset item (`order_items_dataset.csv`).
- Semua analisis berbasis waktu menggunakan kolom **`order_approved_at`** dari `orders_dataset.csv` (setelah di-parse menjadi datetime dan baris dengan `order_approved_at` kosong dihapus).
- Dataset inti yang digunakan untuk revenue & produk adalah tabel hasil merge: **`order_merge = order_items_df.merge(orders_df, on='order_id')`**.

### Q1. Kenaikan Revenue dari tahun ke tahun
1. Buat agregasi tahunan:
   - `revenue_year = sum(price)` dengan grup `order_merge.year`.
2. Hitung **YoY Growth (%)**:
   - `yoy_growth_% = revenue_year.pct_change() * 100` (berdasarkan urutan tahun).
3. Output yang dicari:
   - Tabel `year`, `revenue`, `yoy_growth_%`.
   - Tahun dengan `yoy_growth_%` maksimum.

Kolom yang dipakai:
- `order_merge.year` (turunan dari `order_approved_at`)
- `order_merge.price`
- `order_merge.order_id` (opsional untuk sanity check jumlah order)

### Q2. Dibulan apa penjualan cenderung meningkat
1. Hitung revenue per bulan kalender:
   - grup `order_merge.year_month` + `order_merge.month`.
   - `revenue_month = sum(price)`
2. Hitung **MoM Growth (%)** antar bulan berturut:
   - `mom_growth_% = revenue_month.pct_change() * 100`
3. Tentukan “bulan yang cenderung meningkat” (berbasis tren MoM):
   - Kelompokkan `mom_growth_%` berdasarkan **bulan kalender** (1–12).
   - Hitung metrik:
     - `positive_mom_rate = mean(mom_growth_% > 0)`
     - `avg_mom_growth_% = mean(mom_growth_%)`
   - Pilih bulan kalender dengan ranking:
     - `positive_mom_rate` tertinggi, lalu `avg_mom_growth_%` tertinggi.

Kolom yang dipakai:
- `order_merge.year_month`, `order_merge.month`
- `order_merge.price`

### Q3. Apa 3 product terlaris setiap bulan-nya
Metode:
1. Enrich produk ke nama kategori:
   - merge `order_merge` dengan `products_df` → `product_category_name`.
   - merge lanjutan dengan `product_category_name_translation_df` untuk mendapatkan `product_category_name_english`.
2. Untuk tiap bulan kalender (`year_month`), hitung:
   - `revenue_category = sum(price)` per kategori.
3. Rank kategori per bulan berdasarkan `revenue_category` descending dan ambil **Top 3**.

Kolom yang dipakai:
- `order_merge.year_month`, `order_merge.product_id`, `order_merge.price`
- `products_df.product_id`, `products_df.product_category_name`
- `product_category_name_translation_df.product_category_name`, `product_category_name_english`

### Q4. Persentase customer “sultan” yang menggunakan e-commerce
Definisi RFM (customer level menggunakan `customer_unique_id`):
- **Recency**: jumlah hari sejak pembelian terakhir.
  - `recency_days = (max(order_approved_at) - last_order_date).days`
- **Frequency**: jumlah order unik.
  - `frequency = nunique(order_id)`
- **Monetary**: total belanja.
  - `monetary = sum(price)`

Scoring (kuintil 1–5):
- `Recency_score`: kuintil berdasarkan `recency_days`, dengan **skor lebih tinggi untuk recency yang lebih kecil (lebih baru)**.
- `Frequency_score`: kuintil berdasarkan `frequency`.
- `Monetary_score`: kuintil berdasarkan `monetary`.

Weighted score (sesuai aturan Anda):
- `weighted_score = 5*Monetary_score + 5*Frequency_score + 2*Recency_score`

Definisi Sultan:
- `sultan_flag = 1` jika `weighted_score` berada pada **top 10%** (menggunakan percentile 90).

Output:
- `% Sultan = (jumlah_customer_sultan / jumlah_total_customer) * 100`.

Kolom yang dipakai:
- `order_merge.customer_id` (untuk join ke `customers_df.customer_unique_id`)
- `order_merge.order_approved_at`, `order_merge.order_id`, `order_merge.price`
- `customers_df.customer_unique_id`

### Q5. Kota dengan potensi customer loyal terbesar (RFM + geospatial)
1. Join RFM Sultan ke profil customer:
   - gunakan `customers_df.customer_zip_code_prefix` untuk menyambungkan ke geolocation.
2. Geospatial enrichment:
   - join `customers_df.customer_zip_code_prefix == geo_df.geolocation_zip_code_prefix`.
   - gunakan `geo_df.geolocation_lat` dan `geo_df.geolocation_lng`.
3. Definisi “potensi loyal” per kota:
   - buat label kota: `city = customer_city + ', ' + customer_state`.
   - `sultan_count_city = sum(sultan_flag)`
   - `total_customers_city = count(customers)`
   - `sultan_share_city = sultan_count_city / total_customers_city`
4. Output:
   - Top kota berdasarkan `sultan_share_city`.
   - Map scatter berdasarkan lat/lng dengan marker size proporsional `sultan_count_city`.

Kolom yang dipakai:
- `customers_df.customer_city`, `customers_df.customer_state`, `customers_df.customer_zip_code_prefix`
- `geo_df.geolocation_lat`, `geo_df.geolocation_lng`, `geo_df.geolocation_zip_code_prefix`
- `rfm_df.sultan_flag`

## Analisis Lanjutan (Opsional)

```python
# Q3, Q4, Q5 (Analisis Lanjutan)

# -------------------------
# Q3: Top 3 product per bulan
# -------------------------
products_enriched = products_df.merge(
    product_category_name_translation_df,
    on='product_category_name',
    how='left'
)

# Fallback jika terjemahan bahasa Inggris kosong
products_enriched['product_category_name_english'] = products_enriched['product_category_name_english'].fillna(products_enriched['product_category_name'])

order_with_product = order_merge.merge(
    products_enriched[['product_id', 'product_category_name_english']],
    on='product_id',
    how='left'
)

cat_month_revenue = (
    order_with_product.groupby(['year_month', 'product_category_name_english'], as_index=False)
    .agg(revenue=('price', 'sum'))
)

cat_month_revenue['rank'] = (
    cat_month_revenue.groupby('year_month')['revenue']
    .rank(ascending=False, method='first')
)

top3_rows = cat_month_revenue[cat_month_revenue['rank'] <= 3].copy()

top3_by_month = (
    top3_rows.sort_values(['year_month','rank'])
    .pivot(index='year_month', columns='rank', values='product_category_name_english')
    .rename(columns={1:'top1_category', 2:'top2_category', 3:'top3_category'})
    .reset_index()
    .sort_values('year_month')
)

print('Top 3 kategori per bulan (berdasarkan sum(price))')
print(top3_by_month.to_string(index=False))

# -------------------------
# Q4: RFM + Sultan %
# -------------------------
order_merge_cust = order_merge.merge(
    customers_df[['customer_id', 'customer_unique_id', 'customer_zip_code_prefix', 'customer_city', 'customer_state']],
    on='customer_id',
    how='left'
)
order_merge_cust = order_merge_cust.dropna(subset=['customer_unique_id']).copy()

customer_last_order = order_merge_cust.groupby('customer_unique_id')['order_approved_at'].max()
customer_recency_days = (order_merge_cust['order_approved_at'].max() - customer_last_order).dt.days

customer_frequency = order_merge_cust.groupby('customer_unique_id')['order_id'].nunique()
customer_monetary = order_merge_cust.groupby('customer_unique_id')['price'].sum()

rfm_df = pd.DataFrame({
    'recency_days': customer_recency_days,
    'frequency': customer_frequency,
    'monetary': customer_monetary,
}).reset_index()

# After reset_index(), the customer id column name can differ by pandas version
if 'customer_unique_id' not in rfm_df.columns:
    rfm_df = rfm_df.rename(columns={'index': 'customer_unique_id'})

# Quintile scoring 1..5
# Recency: skor lebih tinggi untuk recency yang lebih kecil (lebih baru)
rfm_df['Recency_score'] = pd.qcut(
    rfm_df['recency_days'],
    q=5,
    labels=[5,4,3,2,1],
    duplicates='drop'
).astype(int)

rfm_df['Frequency_score'] = pd.qcut(
    rfm_df['frequency'],
    q=5,
    labels=[1,2,3,4,5],
    duplicates='drop'
).astype(int)

rfm_df['Monetary_score'] = pd.qcut(
    rfm_df['monetary'],
    q=5,
    labels=[1,2,3,4,5],
    duplicates='drop'
).astype(int)

rfm_df['weighted_score'] = (
    RFM_WEIGHTS['Monetary'] * rfm_df['Monetary_score'] +
    RFM_WEIGHTS['Frequency'] * rfm_df['Frequency_score'] +
    RFM_WEIGHTS['Recency'] * rfm_df['Recency_score']
)

sultan_threshold = rfm_df['weighted_score'].quantile(SULTAN_QUANTILE)
rfm_df['sultan_flag'] = (rfm_df['weighted_score'] >= sultan_threshold).astype(int)

sultan_count = int(rfm_df['sultan_flag'].sum())
sultan_percent = sultan_count / len(rfm_df) * 100

print('\nRFM ringkasan per customer (untuk sultan):')
print(rfm_df[['recency_days','frequency','monetary','Recency_score','Frequency_score','Monetary_score','weighted_score','sultan_flag']].describe())
print('\nDefinisi Sultan: top 10% weighted_score (>= percentile 90)')
print(f'Sultan count: {sultan_count} dari total {len(rfm_df)} customer')
print(f"Persentase customer Sultan: {sultan_percent:.2f}%")

# -------------------------
# Q5: Geospatial + City potential
# -------------------------
city_profile = rfm_df.merge(
    customers_df[['customer_unique_id', 'customer_zip_code_prefix', 'customer_city', 'customer_state']],
    on='customer_unique_id',
    how='left'
)

geo_enriched = city_profile.merge(
    geo_df[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']].drop_duplicates(subset=['geolocation_zip_code_prefix']),
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)

geo_enriched = geo_enriched.dropna(subset=['geolocation_lat','geolocation_lng','customer_city','customer_state']).copy()
geo_enriched['city'] = geo_enriched['customer_city'].astype(str) + ', ' + geo_enriched['customer_state'].astype(str)

city_stats = (
    geo_enriched.groupby('city', as_index=False)
    .agg(
        total_customers_city=('customer_unique_id','nunique'),
        sultan_customers_city=('sultan_flag','sum'),
        sultan_share_city=('sultan_flag', lambda s: s.sum() / s.shape[0])
    )
)

# Tie-break: higher share then higher sultan count
city_stats = city_stats.sort_values(['sultan_share_city','sultan_customers_city'], ascending=False)

top_cities = city_stats.head(10).copy()
print('\nTop 10 city berdasarkan sultan_share_city (RFM + geospatial):')
print(top_cities.to_string(index=False))

best_city = top_cities.iloc[0]['city']
best_share = float(top_cities.iloc[0]['sultan_share_city'])
best_count = int(top_cities.iloc[0]['sultan_customers_city'])
print(f"\nKota dengan potensi customer loyal (RFM) terbesar: {best_city}")
print(f"Sultan share: {best_share:.2%} | Sultan count: {best_count}")

# City coordinates for mapping (approximate by mean lat/lng)
city_coords = (
    geo_enriched.groupby('city', as_index=False)
    .agg(lat=('geolocation_lat','mean'), lng=('geolocation_lng','mean'))
)

city_map_df = top_cities.merge(city_coords, on='city', how='left')

fig = px.scatter_geo(
    city_map_df,
    lat='lat',
    lon='lng',
    size='sultan_customers_city',
    color='sultan_share_city',
    hover_name='city',
    projection='natural earth',
    scope='south america',
    title='Q5 - City potential (RFM Sultan)'
)
fig.show()
```

- Q1: Kenaikan revenue YoY dan tahun dengan YoY growth tertinggi (berdasarkan tabel `yearly_revenue`).
- Q2: Bulan kalender yang cenderung meningkat (berdasarkan `positive_mom_rate` tertinggi lalu `avg_mom_growth_%`).
- Q3: Top 3 kategori produk terlaris untuk setiap `year_month` (tabel `top3_by_month`).
- Q4: Persentase customer “sultan” (top 10% weighted RFM) pada dataset.
- Q5: Kota dengan potensi customer loyal terbesar (top kota berdasarkan `sultan_share_city`) + peta sebaran.
