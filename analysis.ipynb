## Cek Info, Hapun Null Value, & Hapus Outlier

```python
orders_df.info()
```

```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'])
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'])

orders_df.head()
```

```python
orders_df.isna().sum()
```

```python
orders_df.shape
```

```python
orders_df = orders_df.dropna(subset=['order_approved_at'])
print("total rows after", orders_df.shape[0])
```

```python
order_items_df.info()
```

```python
order_items_df.isna().sum()
```

```python
Q3 = order_items_df['price'].quantile(0.75)
Q1 = order_items_df['price'].quantile(0.25)

IQR = Q3 - Q1

order_items_df = order_items_df[order_items_df['price'] <= 1.5 * IQR]
print("Total Length: ", order_items_df.shape[0])
```

```python
order_merge = order_items_df.merge(orders_df, on='order_id')
print("Length: ", len(order_merge))
order_merge.head()
```

```python
sns.histplot(order_merge['price'])
```

```python
order_price_agg = order_merge.groupby(by=order_merge['order_approved_at'].dt.date)['price'].sum()
order_price_agg.head()
```

```python
sns.lineplot(data=order_price_agg)
```

```python
# quartal monthly performace
sales_df = order_merge[['order_id', 'order_item_id', 'price', 'order_approved_at']]
sales_df['quarter'] = sales_df['order_approved_at'].dt.quarter
sales_df['year'] = sales_df['order_approved_at'].dt.year


quarter_sales = (
    sales_df.groupby(by=['quarter', 'year'], as_index=False)
    .agg(total_sales=('price', 'sum'))
)

quarter_sales
```

```python
sns.barplot(data=quarter_sales, x='year', y='total_sales', hue='quarter')
```
